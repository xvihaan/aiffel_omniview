step 6. EEVE-Korean-10.8B(ggml-model-Q5_K_M.gguf) 모델 사용

```python
import ollama

response = ollama.chat(model='EEVE-Korean-10.8B:latest', 
                     messages=[{
                         'role':'user',
                         'content':'미에로화이바 맛은 어때?'}],
                     stream=True)

for chunk in response:
    print(chunk['message']['content'], end='', flush=True)
```

```python
import time
import langchain
from langchain_ollama import OllamaLLM
import logging

# Configure basic logging
logging.basicConfig(level=logging.INFO)

try:
    llm = OllamaLLM(model='EEVE-Korean-10.8B:latest')
    
    # 프롬프트가 잘 정의되어 있는지 확인하는 것이 필요합니다. (모델의 기능에 따라 조정가능)
    start = time.time()
    prompt = ("미에로화이바 맛은 어때?")
    
    response = llm.invoke(prompt)
    print(response)
    print(time.time() - start)
    
except ImportError:
    logging.error("Failed to import Ollama from langchain_community. Is the package installed?")
except Exception as e:
    logging.error(f"An unexpected error occurred: {e}")
```

```python
retriever = GoogleSerperAPIWrapper(k=2)

prompt = ChatPromptTemplate.from_template(
'Context: {context}'
)


def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)


'''chain = (
    {"context": retriever , "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)'''

from langchain_core.runnables import RunnableLambda

# retriever의 run 메서드 사용 (실제 메서드 이름은 확인 필요)
runnable_retriever = RunnableLambda(lambda query: retriever.run(query))

chain = (
    {"context": runnable_retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

result = chain.invoke(f"{미에로화이바} 맛은 어때?")
print(result)
```
