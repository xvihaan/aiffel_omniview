## 데이터 증식(Data Argumentation)

```python
import os
from keras.preprocessing.image import ImageDataGenerator

# 데이터 경로 설정
rootPath = '/content/drive/MyDrive/Datasets/image'

# 스케일링 + 데이터 증식 속성 설정
trainImageGenerator = ImageDataGenerator(rescale=1./255,
                                         horizontal_flip=True,
                                         vertical_flip=True,
                                         shear_range=0.5,
                                         brightness_range=[0.5, 1.5],
                                         zoom_range=0.2,
                                         width_shift_range=0.1,
                                         height_shift_range=0.1,
                                         rotation_range=30,
                                         fill_mode='nearest')

# 스케일링 설정
testImageGenerator = ImageDataGenerator(rescale=1./255)

# 데이터 스케일링 및 증식
trainGen = trainImageGenerator.flow_from_directory(
    os.path.join(rootPath, 'Training'),
    target_size = (64, 64), # 이미지 사이즈를 64, 64로 학습 - 모델 input_shape에도 이걸 적어줘야 함
    class_mode = 'categorical' # 다중 분류일 때는 categorical
)
# # 데이터 제너레이터 설정 시 label_mode를 'int'로 설정하여 정수 형태의 레이블 반환
# trainGen = trainDataGen.flow_from_directory(train_dir, target_size=(64, 64), batch_size=32, class_mode='sparse')
# testGen = testDataGen.flow_from_directory(test_dir, target_size=(64, 64), batch_size=32, class_mode='sparse')

testGen = testImageGenerator.flow_from_directory(
    os.path.join(rootPath, 'Validation'),
    target_size = (64, 64),
    class_mode = 'categorical'
)
```

```python
# 모델층 얕게 수정

from keras.models import Sequential
from keras import layers
import numpy as np

np.random.seed(77)

model = Sequential()

model.add(layers.InputLayer(input_shape=(64, 64, 3)))
model.add(layers.Conv2D(filters=16, kernel_size=3, padding='same', activation='relu'))
model.add(layers.MaxPool2D(pool_size=(2, 2)))
model.add(layers.Dropout(rate=0.3))

model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
from keras.utils import plot_model

plot_model(model, 'model.png' ,show_shapes=True)
```

## 모델 설정하기

```python
model.compile(optimizer = 'adam',
              loss = 'categorical_crossentropy',
              metrics = ['acc'])
```

## 모델 학습하기

```python
history_77 = model.fit(trainGen, epochs = 10,
                    validation_data = testGen)
```

## 학습 결과 분석

```python
import matplotlib.pyplot as plt

his_dict = history_77.history
loss = his_dict['loss']
val_loss = his_dict['val_loss'] # 검증 데이터가 있는 경우 ‘val_’ 수식어가 붙습니다.

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize = (10, 5))

# 훈련 및 검증 손실 그리기
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color = 'blue', label = 'train_loss')
ax1.plot(epochs, val_loss, color = 'orange', label = 'val_loss')
ax1.set_title('train and val loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()

acc = his_dict['acc']
val_acc = his_dict['val_acc']

# 훈련 및 검증 정확도 그리기
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, acc, color = 'blue', label = 'train_acc')
ax2.plot(epochs, val_acc, color = 'orange', label = 'val_acc')
ax2.set_title('train and val acc')
ax2.set_xlabel('epochs')
ax2.set_ylabel('loss')
ax2.legend()

plt.show()
```

## 모델 평가하기

```python
scores = model.evaluate(testGen)
print("%s : %.2f%%" % (model.metrics_names[1], scores[1] * 100))
```

## 사전학습 모델을 활용해서 예측 정확도 높이기

*   학습 데이터, 테스트 데이터 그대로 사용
*   첫 번째 모델 VGG16 사용, GlobalAveragePooling2D() + Flatten() + Dense(256) + Dense(64) + Dense(1)
*   두 번째 모델 ResNet50V2 사용, GlobalAveragePooling2D() + Flatten() + Dense(256) + Dense(64) + Dense(1)

### 첫 번째 모델 : VGG16

```python
from keras.models import Sequential
from keras.applications import VGG16 # 사전학습 모델
from keras.layers import Conv2D, GlobalAveragePooling2D, Dropout, Flatten, Dense
# GlobalAveragePooling2D : 전체 데이터의 평균값으로 풀링
from keras.optimizers import Adam

# 전이학습(사전학습) 모델 가져오기
pre_trained_vgg = VGG16(input_shape = (64, 64, 3), # 내 데이터의 input_shape을 줘야 함
                        include_top = False,
                        weights = 'imagenet') # 가중치

# 분류모델 합치기
model_vgg = Sequential()
model_vgg.add(pre_trained_vgg) # 기존에 만든 모델에 사전학습 모델 끼우기
model_vgg.add(layers.GlobalAveragePooling2D()) # 풀링으로 특성 추출

# 파인튜닝
model_vgg.add(layers.Flatten())
model_vgg.add(layers.Dense(64, activation='relu'))
model_vgg.add(layers.Dense(10, activation='softmax'))
```

```python
model_vgg.summary()
```

```python
model_vgg.compile(optimizer = Adam(1e-4), loss = 'categorical_crossentropy', metrics = ['acc'])
history_vgg = model_vgg.fit(trainGen, epochs = 10,
                    validation_data = testGen)
```

```python
import matplotlib.pyplot as plt

his_dict = history_vgg.history
loss = his_dict['loss']
val_loss = his_dict['val_loss'] # 검증 데이터가 있는 경우 ‘val_’ 수식어가 붙습니다.

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize = (10, 5))

# 훈련 및 검증 손실 그리기
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color = 'blue', label = 'train_loss')
ax1.plot(epochs, val_loss, color = 'orange', label = 'val_loss')
ax1.set_title('train and val loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()

acc = his_dict['acc']
val_acc = his_dict['val_acc']

# 훈련 및 검증 정확도 그리기
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, acc, color = 'blue', label = 'train_acc')
ax2.plot(epochs, val_acc, color = 'orange', label = 'val_acc')
ax2.set_title('train and val acc')
ax2.set_xlabel('epochs')
ax2.set_ylabel('loss')
ax2.legend()

plt.show()
```

```python
# 첫 번째 모델 : VGG16 정확도
model_vgg.evaluate(testGen)
```

### 두 번째 모델 : ResNet50V2
